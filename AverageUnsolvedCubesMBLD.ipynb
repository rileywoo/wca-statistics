```python
import pandas as pd
import numpy as np
```

```python
# The average number of unsolved cubes in multi-BLD attempts (non DNF attempts)
results_path = 'WCA_export_Results.tsv'
results_df = pd.read_table(results_path)[['personId', 'eventId', 'value1', 'value2', 'value3']]
results_df.head()
```

```python
mbld_results = results_df[results_df['eventId'] == '333mbf'][['personId', 'value1', 'value2', 'value3']]
```

```python
mbld_results['combined'] = mbld_results[['value1', 'value2', 'value3']].values.tolist()
```

```python
mbld_results = mbld_results[['personId', 'combined']]
```

```python
mbld_results = mbld_results.groupby('personId').agg(sum)
```

```python
def calc_average_unsolved(x):
    def map_helper(result):
        return int(str(result)[-2:])
    valid_results = list(filter(lambda val: val > 0, x['combined'][0]))
    num_unsolved_list = list(map(map_helper, valid_results))
    return np.mean(num_unsolved_list)

num_unsolved_series = mbld_results.groupby('personId').apply(calc_average_unsolved)
mbld_results['mystery'] = num_unsolved_series
final_df = mbld_results[['mystery']]
```

```python
final_df = final_df.reset_index().sort_values(by='mystery', ascending=False)
persons_path = "WCA_export_Persons.tsv"
persons_df = pd.read_table(persons_path)
final_df = final_df.merge(persons_df, how='left', left_on='personId', right_on='id')[['name', 'personId', 'mystery']]
```
